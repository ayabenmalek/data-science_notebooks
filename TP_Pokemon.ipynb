## L'univers des Pokémons

Un Pokémon est un animal issu du monde des jeux vidéo. Chaque Pokémon possède des caractéristiques qui lui sont propres, à savoir son type (Pokémon d'herbe, Pokémon de feu ... ) lui donnant alors des facultés spéciales. Ainsi, le Pokémon nommé `Dracofeu`, de type `Feu` est capable de cracher du feu. De plus, on note l'existence de Pokémons légendaires très rares et très puissants (Pokémon diamant, Pokémon platine...).

Les Pokémons sont également de différentes générations faisant référence à l'évolution de l'univers des Pokémons. Les Pokémons de première génération sont ceux ayant pris naissance au tout début de la création de l'univers (création de 151 Pokémons) pour arriver à la huitième génération avec plus de 800 Pokémons (2020)!

Dans cet univers viennent ensuite les dresseurs de Pokémons. Les dresseurs les collectionnent et les élèvent dans le but de les faire combattre entre eux jusqu'au KO. Lors d'un combat de Pokémons, opposant deux dresseurs, l'un choisit l'animal qui sera le premier à entrer dans l'arène. En fonction de ce premier Pokémon, le second dresseur choisit un animal dans sa collection qui sera en mesure de battre le premier Pokémon. Si le premier Pokémon présent dans l'arène est de type feu, il y a de fortes chances que le Pokémon choisi par l'adversaire soit de type eau afin de contrer les attaques du premier.

Cela nécessite pour le dresseur d'avoir une bonne connaissance de chaque Pokémon pour choisir celui qui pourra devenir vainqueur du combat. Pour l'aider dans son choix, les caractéristiques des Pokémons sont répertoriées dans un document appelé Pokédex.

## Agrégation des données avec le Pokédex

Cette agrégation va nous permettre d'avoir une vue centralisée des données et de permettre une analyse plus approfondie.

#### - Créer une liste comportant :
- Les numéros des Pokémons
- Le nombre de combats menés
- Le nombre de combats gagnés
- Le pourcentage de victoires par rapport au nombre de combats menés

#### - Créer un nouveau Pokedex comportant les noms de Pokemons et leur victoire


```python
# Align the Series based on their indices
winners_aligned, nbr_combats_aligned = winners.align(nbr_combats, join='outer', fill_value=0)

new_pokedex = pd.DataFrame({
    'NUMERO' : nbr_combats_aligned.index,
    'Nbr_combats' : nbr_combats_aligned.values,
    'Nbr_victoires' : winners_aligned.values,
    'Pourcentage_victoires' : (winners_aligned.values / nbr_combats_aligned.values * 100)
})

new_pokedex = pd.merge(carac_trans, new_pokedex, on='NUMERO')
print(new_pokedex)
```

# preparing the model

```python
y = combats_full["Winner"]
x = combats_full.drop(["Winner", "First_pokemon", "Second_pokemon", "NUMERO", "NOM", "NOM_cand"], axis=1)
print(x.info())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

## Random forest (no scale needed )

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

y_pred = rf_model.predict(x_test)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Récupérer l’importance
importances = rf_model.feature_importances_

# Associer avec les noms des colonnes
feat_importances = pd.Series(importances, index=x_train.columns)

# Trier et afficher
feat_importances = feat_importances.sort_values(ascending=False)
print(feat_importances)

# Visualisation
feat_importances.plot(kind='bar', figsize=(10,5))
plt.title("Importance des features dans Random Forest")
plt.show()
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)

y_pred = dt_model.predict(x_test)
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

# tester sur le dataset tests

```python
import pandas as pd

def prepare_features(tests, carac_trans, X_train_columns, scaler=None):
    """
    Prépare les features pour prédire sur le dataset `tests`
    en fusionnant avec les caractéristiques des Pokémons.

    - tests : DataFrame avec colonnes ["First_pokemon", "Second_pokemon"]
    - carac_trans : DataFrame contenant les caractéristiques (NUMERO, stats, types, etc.)
    - X_train_columns : colonnes utilisées pour entraîner le modèle
    - scaler : StandardScaler déjà fit (optionnel)

    Retourne :
    - X_test_final prêt pour prédiction
    - tests (copie) enrichi
    """

    # 1 Merge des stats du premier Pokémon
    test_full = tests.merge(
        carac_trans,
        left_on="First_pokemon",
        right_on="NUMERO",
        suffixes=('', '_adv')
    ).drop(["NUMERO", "NOM"], axis=1)

    # 2 Merge des stats du second Pokémon
    test_full = test_full.merge(
        carac_trans,
        left_on="Second_pokemon",
        right_on="NUMERO",
        suffixes=('', '_cand')
    ).drop(["NUMERO", "NOM"], axis=1)

    # 3 Supprimer les colonnes qui ne servent pas directement au modèle
    X_test_final = test_full.drop(["First_pokemon", "Second_pokemon"], axis=1)

    # 4 Aligner les colonnes avec celles vues à l’entraînement
    X_test_final = X_test_final.reindex(columns=X_train_columns, fill_value=0)

    # 5 Scaling si fourni
    if scaler is not None:
        X_test_final = scaler.transform(X_test_final)

    return X_test_final
```

```python
# On suppose que tu as déjà X_train et scaler
X_train_columns = x_train.columns.tolist()

# Préparer les features pour tests
X_test_final = prepare_features(tests, carac_trans, X_train_columns, scaler)

# Prédiction
tests['Winner'] = rf_model.predict(X_test_final)

print(tests)
```
